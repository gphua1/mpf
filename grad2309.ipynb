```python
from __future__ import division
import numpy as np
from numpy.linalg import norm
from numpy.random import rand
from numpy.random import randint
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy.matlib
import PIL.Image
from scipy.optimize import fmin_l_bfgs_b as minimize
import scipy.misc
from scipy.sparse import identity
import matplotlib.pyplot as pp


def mpfcost(w, b, data):
    a=b


    w=w.reshape(64,64)

    X=data
    print(X.shape)


    eps=0.00000000001
    delt=1.0-2*X

    costco =np.exp(np.multiply(delt,np.dot(w, X)))
    deltA=delt
    deltB=delt.T

    print(deltA.shape)
    print(deltB.shape)

    cost = eps*np.sum(costco)
    Wgrad = eps*(np.dot(deltA,X.T)+np.dot(X,deltB))*cost


    grad=Wgrad.ravel()

    print(cost)

    return cost,grad

def symmetra(n):

    W=np.zeros((n,n))
    for i in range (n):
        for j in range (n):

            if i<j:
                W[i,j]=np.random.normal(0,1)
                W[j,i]=W[i,j]


    return W

if __name__ == "__main__":

    patches = np.load('DataX2309.dat.npy')
    c=0
    #W=symmetra(64)
    W=np.random.rand(64,64)
    ww=W.ravel()

    opttheta, cost, messages = minimize(mpfcost, ww, fprime=None, maxiter=1000,args =(c, patches))

    Wmpf = opttheta.reshape(64,64)
    x, y = Wmpf.shape
    Wreal = np.load('Wreal2309.dat.npy')
    mse = np.sum(np.square(Wreal - Wmpf)) / (x * y)
    print(mse)
    WmpfLine = np.ravel(Wmpf)
    WrealLine = np.ravel(Wreal)
    pp.plot(WmpfLine[200:220])
    pp.plot(WrealLine[200:220])
    pp.show()



```
